```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import pickle
```

```python
data = np.random.randn(1000, 4)
df = pd.DataFrame(data)
df.columns = ["Tempo Fixo", "NPOP = 10, NGEN = 50", "NPOP = 20, NGEN = 100", "NPOP = 40, NGEN = 200"]
```

```python
df.plot.hist()
```

```python
tempo_fixo = pd.DataFrame()
ga_cfg_1 = pd.DataFrame()
ga_cfg_2 = pd.DataFrame()
ga_cfg_3 = pd.DataFrame()
with open("results/corridor/tempo_fixo/vehicles.pickle", "rb") as arq:
    tempo_fixo = pickle.load(arq)
with open("results/corridor/ga_cfg_1/vehicles.pickle", "rb") as arq:
    ga_cfg_1 = pickle.load(arq)
with open("results/corridor/ga_cfg_2/vehicles.pickle", "rb") as arq:
    ga_cfg_2 = pickle.load(arq)
with open("results/corridor/ga_cfg_3/vehicles.pickle", "rb") as arq:
    ga_cfg_3 = pickle.load(arq)
```

```python
configs_ga_corridor = [tempo_fixo, ga_cfg_1, ga_cfg_2 ,ga_cfg_3]
df_tempo_viagem_ga_corridor = pd.DataFrame([d["travel_time"] for d in configs_ga_corridor]).transpose()
df_tempo_viagem_ga_corridor.columns = ["Tempo Fixo", "NPOP = 10, NGEN = 50", "NPOP = 20, NGEN = 100", "NPOP = 40, NGEN = 200"]
df_tempo_viagem_ga_corridor.plot.hist(bins=20,
                                      alpha=0.3)
plt.ylabel("")
plt.xlabel("Tempo de Viagem (s)")
```

```python
df_tempo_viagem_ga_corridor.describe()
```

```python
configs_ga_corridor = [tempo_fixo, ga_cfg_1, ga_cfg_2 ,ga_cfg_3]
df_tempo_parado_ga_corridor = pd.DataFrame([d["stopped_time"] for d in configs_ga_corridor]).transpose()
df_tempo_parado_ga_corridor.columns = ["Tempo Fixo", "NPOP = 10, NGEN = 50", "NPOP = 20, NGEN = 100", "NPOP = 40, NGEN = 200"]
df_tempo_parado_ga_corridor.plot.hist(bins=20,
                                      alpha=0.3)
plt.ylabel("")
plt.xlabel("Tempo Parado (s)")
```

```python
df_tempo_parado_ga_corridor.describe()
```

```python
tempo_fixo_dem400 = pd.DataFrame()
tempo_fixo_dem600 = pd.DataFrame()
tempo_fixo_dem800 = pd.DataFrame()
tempo_fixo_dem1000 = pd.DataFrame()
scoga_dem400 = pd.DataFrame()
scoga_dem600 = pd.DataFrame()
scoga_dem800 = pd.DataFrame()
scoga_dem1000 = pd.DataFrame()
itlc_dem400 = pd.DataFrame()
itlc_dem600 = pd.DataFrame()
itlc_dem800 = pd.DataFrame()
itlc_dem1000 = pd.DataFrame()
with open("results/corridor/tempo_fixo_demanda_400/vehicles.pickle", "rb") as arq:
    tempo_fixo_dem400 = pickle.load(arq)
with open("results/corridor/tempo_fixo_demanda_600/vehicles.pickle", "rb") as arq:
    tempo_fixo_dem600 = pickle.load(arq)
with open("results/corridor/tempo_fixo_demanda_800/vehicles.pickle", "rb") as arq:
    tempo_fixo_dem800 = pickle.load(arq)
with open("results/corridor/tempo_fixo_demanda_1000/vehicles.pickle", "rb") as arq:
    tempo_fixo_dem1000 = pickle.load(arq)
with open("results/corridor/scoga_demanda_1000/vehicles.pickle", "rb") as arq:
    scoga_dem1000 = pickle.load(arq)
with open("results/corridor/scoga_demanda_800/vehicles.pickle", "rb") as arq:
    scoga_dem800 = pickle.load(arq)
with open("results/corridor/scoga_demanda_600/vehicles.pickle", "rb") as arq:
    scoga_dem600 = pickle.load(arq)
with open("results/corridor/scoga_demanda_400/vehicles.pickle", "rb") as arq:
    scoga_dem400 = pickle.load(arq)
with open("results/corridor/itlc_demanda_400/vehicles.pickle", "rb") as arq:
    itlc_dem400 = pickle.load(arq)
with open("results/corridor/itlc_demanda_600/vehicles.pickle", "rb") as arq:
    itlc_dem600 = pickle.load(arq)
with open("results/corridor/itlc_demanda_800/vehicles.pickle", "rb") as arq:
    itlc_dem800 = pickle.load(arq)
with open("results/corridor/itlc_demanda_1000/vehicles.pickle", "rb") as arq:
    itlc_dem1000 = pickle.load(arq)
```

```python
viagem_tempo_fixo = [df.describe().loc["mean"]["travel_time"] for df in
                     [tempo_fixo_dem400, tempo_fixo_dem600, tempo_fixo_dem800, tempo_fixo_dem1000]]
viagem_scoga = [df.describe().loc["mean"]["travel_time"] for df in
                [scoga_dem400, scoga_dem600, scoga_dem800, scoga_dem1000]]
viagem_itlc = [df.describe().loc["mean"]["travel_time"] for df in
                [itlc_dem400, itlc_dem600, itlc_dem800, itlc_dem1000]]
```

```python
parado_tempo_fixo = [df.describe().loc["mean"]["stopped_time"] for df in
                     [tempo_fixo_dem400, tempo_fixo_dem600, tempo_fixo_dem800, tempo_fixo_dem1000]]
parado_scoga = [df.describe().loc["mean"]["stopped_time"] for df in
                [scoga_dem400, scoga_dem600, scoga_dem800, scoga_dem1000]]
parado_itlc = [df.describe().loc["mean"]["stopped_time"] for df in
                [itlc_dem400, itlc_dem600, itlc_dem800, itlc_dem1000]]
```

```python
x = np.arange(4)
larg = 0.2
fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(x - 0.2, viagem_tempo_fixo, larg, label="Tempo Fixo")
ax.bar(x, viagem_itlc, larg, label="ITLC")
ax.bar(x + 0.2, viagem_scoga, larg, label="SCOGA")
ax.set_xticks(x)
ax.set_xticklabels(["400", "600", "800", "1000"])
ax.set_xlabel("Demanda Base (veh/h)")
ax.set_ylabel("Tempo de Viagem Médio (s)")
plt.legend()
plt.tight_layout()
```

```python
x = np.arange(4)
larg = 0.2
fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(x - 0.2, parado_tempo_fixo, larg, label="Tempo Fixo")
ax.bar(x, parado_itlc, larg, label="ITLC")
ax.bar(x + 0.2, parado_scoga, larg, label="SCOGA")
ax.set_xticks(x)
ax.set_xticklabels(["400", "600", "800", "1000"])
ax.set_xlabel("Demanda Base (veh/h)")
ax.set_ylabel("Tempo Parado Médio (s)")
plt.legend()
plt.tight_layout()
```
